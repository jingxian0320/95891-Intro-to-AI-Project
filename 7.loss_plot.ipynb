```
import re
import matplotlib.pyplot as plt
```

```
LOG_PATH = "/content/drive/My Drive/Intro to AI Project/model/logs/"
```

```
def extract_loss(file_name):
    result = []
    log_file = open(LOG_PATH+file_name,"r")
    lines = log_file.readlines()
    for line in lines:
        m = re.search(r".+avg=([0-9\.]+)$", line)
        if (m):
            result.append(float(m.group(1)))
    return result
```

```
loss_sample5 = extract_loss("gpt2-sample5-355M.log")
loss_split10 = extract_loss("gpt2-split10-355M.log")
print (len(loss_sample5))
print (len(loss_split10))
```

```
plt.plot(loss_sample5, label="20%")
plt.plot(loss_split10, label="10 folds")
# Add legend
plt.legend(loc='lower left')
# Add title and x, y labels
plt.title("Training Loss in Fine-tuning", fontsize=16, fontweight='bold')
plt.xlabel("Number of Steps")
plt.ylabel("Avg Loss")
plt.show()
```
